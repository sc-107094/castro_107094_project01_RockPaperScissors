```python
def main():
    win = GraphWin("RockPaperScissors(TEST)", 600, 600, autoflush = False)
    win.setCoords(0, 0, 6, 6)
    win.setBackground("gray")
    
#************************ GUI Intro Set-up ********************************************    
    
    labels = ["Rock", "Paper", "Scissors"]
    
    
    for i in range(3):
        choiceSquare = Rectangle(Point(0 + (i * 2), 0), Point(2 + (i * 2), 2))
        choiceLabels = Text(Point(1 + (i *2), 1), labels[i])
        choiceSquare.setOutline("black")
        choiceSquare.setFill("white")
        choiceSquare.draw(win)
        choiceLabels.draw(win)

        upperSquare = Rectangle(Point(0 + (i * 2), 4), Point(2 + (i * 2), 6))
        mark = Text(Point(1 + (i * 2), 5), "?")
        mark.setTextColor("gold")
        mark.setSize(36)
        upperSquare.setOutline("white")
        upperSquare.setFill("black")
        upperSquare.draw(win)
        mark.draw(win)
        
        
        
    playerLabel = Text(Point(0.5, 2.2), "Player")
    playerLabel.setSize(20)
    playerLabel.setTextColor("aquamarine")
    playerLabel.draw(win)
    
    compLabel = Text(Point(5.6, 3.7), "CPU")
    compLabel.setSize(20)
    compLabel.setTextColor("red")
    compLabel.draw(win)
    
#         choiceLabelRock = Text(Point(1 + 0, 1), "Rock")
#         choiceLabelPaper = Text(Point(1 + 2, 1), "Paper")
#         choiceLabelScissors = Text(Point(1 + 4, 1), "Scissors")
        
#         choiceLabelRock.draw(win)
#         choiceLabelPaper.draw(win)
#         choiceLabelScissors.draw(win)
        
#     for i in range(3):
#         upperSquare = Rectangle(Point(0 + (i * 2), 4), Point(2 + (i * 2), 6))
#         mark = Text(Point(1 + (i * 2), 5), "?")
#         mark.setTextColor("gold")
#         mark.setSize(36)
#         upperSquare.setOutline("white")
#         upperSquare.setFill("black")
#         upperSquare.draw(win)
#         mark.draw(win)
        
#********** Mouse Detection ***************************************************************  
        
        
    alert = Text(Point(3, 3), "Make your Choice!")   
    alert.draw(win)
        
    playerChoice = win.getMouse() 
    
    while playerChoice.getY() > 2:
        alert.setText("Choose from the bottom squares!")
        playerChoice = win.getMouse() 
    
        
    if(playerChoice.getX() <= 2):
        alert.setText("You chose Rock!") #1
        win.getMouse()
    elif(playerChoice.getX() >= 2 and playerChoice.getX() <= 4):
        alert.setText("You chose Paper!") #2
        win.getMouse()
    else:
        alert.setText("You chose Scissors!") #3
        win.getMouse()
        
#*********************************************************************************************
    
    alert.setText("Click again to exit!")
    alert.setTextColor("green")
    alert.setSize(24)
    win.getMouse()
    win.close()
    
main()
```
